```python
                         # 2. UNIVARIATE ANALYSIS
```

```python
#Distribution of Sales,Profit,Quantity:



# a) Distribution of Sales:

plt.figure(figsize=[6,5])
plt.hist(data['Sales'],bins=10,color = 'violet')
plt.title("Sales Distribution in Histogram")
plt.xlabel("Sales")
plt.ylabel("Count")
plt.show()
plt.title("Sales Distribution in Boxplot")
sns.boxplot(x = data['Sales'],color = 'Grey')
plt.show()
```

```python
# b) Distribution of Profit:

plt.figure(figsize=[6,5])
plt.hist(data['Profit'],bins=10,color = 'Yellow')
plt.title("Profit Distribution in Histogram")
plt.xlabel("Profit")
plt.ylabel("Count")
plt.show()
plt.title("Profit Distribution in Boxplot")
sns.boxplot(x = data['Profit'],color = 'Grey')
plt.show()
```

```python
# a) Distribution of Qunatity:

plt.figure(figsize=[6,5])
plt.hist(data['Quantity'],bins=10)
plt.title("Quantity Distribution in Histogram")
plt.xlabel("Quantity")
plt.ylabel("Count")
plt.show()
plt.title("Quantity Distribution in Boxplot")
sns.boxplot(x = data['Quantity'],color = 'Grey')
plt.show()
```

```python
  # INFERENCE :

# Find the Most frequent categories/subcategories.

mostFrequentCategory = data["Category"].mode()
print("Most Frequent Category",mostFrequentCategory)
mostFrequentSubCategory = data["SubCategory"].mode()
print("Most Frequent SubCategory",mostFrequentSubCategory)
```

```python
                       # 3. BI-VARIATE ANALYSIS
```

```python
# Compare Sales Vs Profit

sns.scatterplot(x= data['Sales'],y=data['Profit'])
plt.show()
```

```python
# Compare Profit by region:

sns.boxplot(x=data['Region'],y=data['Profit'])
plt.show()
```

```python
# Compare Monthly Sales Trend :

data_grouped = data.groupby("Month",as_index = False)["Sales"].sum()
print(data_grouped)
sns.lineplot(data = data_grouped , x = 'Month', y = 'Sales',marker = "o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
                       # 4. MULTI-VARIATE ANALYSIS
```

```python
# Heatmap correlation between numerical columns (Sales,Profit,Quantity):

d2 = data.tail(15)
Numerical_col = d2[['Sales','Profit','Quantity']].corr()
sns.heatmap(Numerical_col,annot = True,cmap='inferno')
plt.show()
```

```python
# GROUPED ANALYSIS : Sales by Region & Category (pivot + heatmap).


# Pivot: Region vs Category with Sales
pivot_table = data.pivot_table(
    values="Sales",
    index="Region",
    columns="Category",
    aggfunc="sum",
    fill_value=0
)

print(pivot_table)

# Heatmap of pivoted data
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="Blues")
plt.title("Sales by Region & Category")
plt.show()
```
